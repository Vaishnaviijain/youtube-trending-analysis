```python
!pip install pandas requests
import requests
import pandas as pd

API_KEY = 'AIzaSyDBJmd8nQa_bMsuBCU_Hgd7mLd17JI1rBE'
BASE_URL = 'https://www.googleapis.com/youtube/v3/videos'

params = {
    'part': 'snippet,statistics,contentDetails',
    'chart': 'mostPopular',
    'regionCode': 'IN',  # You can change to 'US' or any other country
    'maxResults': 50,
    'key': API_KEY
}

response = requests.get(BASE_URL, params=params)
data = response.json()
video_list = []

for item in data['items']:  # <-- this is the correct key
    video = {
        'title': item['snippet']['title'],
        'channel': item['snippet']['channelTitle'],
        'category': item['snippet']['categoryId'],
        'publish_time': item['snippet']['publishedAt'],
        'tags': item['snippet'].get('tags', []),
        'views': int(item['statistics'].get('viewCount', 0)),
        'likes': int(item['statistics'].get('likeCount', 0)),
        'comments': int(item['statistics'].get('commentCount', 0)),
        'duration': item['contentDetails']['duration']
    }
    video_list.append(video)

df = pd.DataFrame(video_list)
df.head()
df['publish_time'] = pd.to_datetime(df['publish_time'])              # Convert to datetime
df['publish_day'] = df['publish_time'].dt.day_name()                 # Extract day of the week
df['publish_hour'] = df['publish_time'].dt.hour                      # Extract hour (023)

df['tags'] = df['tags'].apply(lambda x: x if isinstance(x, list) else [])  # Ensure it's always a list
!pip install isodate

import isodate

def duration_to_minutes(duration):
    try:
        d = isodate.parse_duration(duration)
        return d.total_seconds() / 60
    except:
        return None

df['duration_mins'] = df['duration'].apply(duration_to_minutes)

df[['views', 'likes', 'comments']] = df[['views', 'likes', 'comments']].fillna(0)
df['views'] = df['views'].astype(int)
df['likes'] = df['likes'].astype(int)
df['comments'] = df['comments'].astype(int)
# Top 10 channels in trending
df['channel'].value_counts().head(10)
# Upload count by day of the week
df['publish_day'].value_counts()
# Average views by upload hour
df.groupby('publish_hour')['views'].mean().sort_values(ascending=False)
# Most common tags (top 10)
from collections import Counter
all_tags = [tag for tags in df['tags'] for tag in tags]
common_tags = Counter(all_tags).most_common(10)
print(common_tags)
# Top 5 videos by likes
df.sort_values(by='likes', ascending=False).head(5)[['title', 'likes']]
# Top 5 videos by comments
df.sort_values(by='comments', ascending=False).head(5)[['title', 'comments']]
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")  # Makes your charts look nice
# Get top 5 categories (still showing category ID for now)
top_categories = df['category'].value_counts().head(5)

# Plot
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Top 5 Video Categories (by Category ID)')
plt.xlabel('Category ID')
plt.ylabel('Number of Videos')
plt.show()
df.groupby('publish_hour')['views'].mean().plot(kind='line', marker='o')
plt.title('Average Views by Upload Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Avg Views')
plt.grid(True)
plt.show()
sns.boxplot(data=df, x='publish_day', y='views')
plt.title('Views by Day of the Week')
plt.xticks(rotation=45)
plt.show()
sns.scatterplot(data=df, x='likes', y='views')
plt.title('Views vs Likes')
plt.xlabel('Likes')
plt.ylabel('Views')
plt.show()
correlation = df[['views', 'likes', 'comments']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

!pip install wordcloud

from wordcloud import WordCloud

tag_string = ' '.join([tag for tags in df['tags'] for tag in tags])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(tag_string)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Tags')
plt.show()
```
